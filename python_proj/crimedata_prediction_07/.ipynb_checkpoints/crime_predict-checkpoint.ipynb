```python
''' Deal with missing values. First, calculate the percentage of
missing values for every column, and plot them as a bar chart'''

null_vals = sample.isnull().sum()/len(sample)*100
null_vals = pd.DataFrame(null_vals)
null_vals.reset_index(inplace = True)
null_vals.columns = ["Feature","Percent missing"]
plt.figure(figsize = (8,6))
plt.xticks(rotation=45)
sns.barplot(x = "Feature",y ="Percent missing",data = null_vals)
#we have some NaN to deal with
```

```python
'''You can see that several of the columns or features are "object" type
These need to be changed to category before we can convert those to 
mappings and numbers'''
#1 (a) Define a function to convert object types and string types to category type

def str_to_cat(my_df):
    for p,q in my_df.items(): #my_df.items() is a generator in Python
        if is_string_dtype(q): 
            my_df[p] = q.astype('category').cat.as_ordered()
    return my_df
```

```python
# change string to datetime format 

sample["DateTime"] = pd.to_datetime(sample["Offense Start DateTime"])
```

```python
#check date format in new colunm
sample.loc[:,"DateTime"]
```

```python
sample.head()
```

```python
#change date into string to avoid gaps in x-axis
sample['yearMonth'] = pd.to_datetime(sample['DateTime'], format='%m-%Y').dt.strftime('%Y%m')


display(sample.head(10))
sample.info()
```

```python
l = sample["yearMonth"].values.tolist()
len(l)
```

```python
#use regular expression to filter the data we need
import re

pattern = re.compile(r'201[2-9]')
pattern2 = re.compile(r'202[0-2]')
d = {}
d2 = {}

for yearMonth in l:
    try:
        match = pattern.match(yearMonth)
        match2 = pattern2.match(yearMonth)
        if match:
            if yearMonth not in d:
                d[yearMonth] = 0
            d[yearMonth] += 1
        elif match2:
            if yearMonth not in d2:
                d2[yearMonth] = 0
            d2[yearMonth] += 1
            
    except:
        continue
        
display(d)
display(d2)

del d2['202212']
del d2['202211']

display(d2)
```

```python
lists = sorted(d.items()) # sorted by key, return a list of tuples
lists2 = sorted(d2.items())

x, y = zip(*lists) # unpack a list of pairs into two tuples
x2, y2 = zip(*lists2)

f, figure = plt.subplots(1)
figure.plot(x, y)
figure.plot(x2, y2)

figure.set_ylim(ymin=0)

#figure.locator_params(axis='x', nbins=10)
every_nth = 12
for n, label in enumerate(figure.xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)

```

```python
xnp = list(range(96))
xnp2 = list(range(34))
xnpPred = list(range(96+36))
```

```python
# Import the packages and classes needed in this example:
import numpy as np
from sklearn.linear_model import LinearRegression

# Create a numpy array of data:
xnp = np.array(xnp).reshape((-1, 1))
ynp = np.array(y)

xnp2 = np.array(xnp2).reshape((-1, 1))
ynp2 = np.array(y2)

# Create an instance of a linear regression model and fit it to the data with the fit() function:
model = LinearRegression().fit(xnp, ynp)

model2 = LinearRegression().fit(xnp2, ynp2)

# The following section will get results by interpreting the created instance: 

# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
r_sq = model.score(xnp, ynp)
print('coefficient of determination:', r_sq)

r_sq2 = model2.score(xnp2, ynp2)
print('coefficient of determination2:', r_sq2)

# Print the Intercept:
b = model.intercept_
print('intercept:', b)

b2 = model2.intercept_
print('intercept2:', b2)

# Print the Slope:
a = model.coef_
print('slope:', a)

a2 = model2.coef_
print('slope:', a2)

# Predict a Response and print it:
y_pred = model.predict(xnp)
print('best fit line', y_pred, sep='\n')
print(len(y_pred))

for i in range(36):
    y_pred = np.append(y_pred, (y_pred[-1]+a))
    
print('Predicted response:', y_pred, sep='\n')
print(len(y_pred))
```
